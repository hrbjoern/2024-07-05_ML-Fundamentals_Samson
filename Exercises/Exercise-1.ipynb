```python
#import libraries
import pandas as pd
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor

#load data
iowa_file_path = "../data/Housing Prices Competition/train.csv"
home_data = pd.read_csv(iowa_file_path)

# show data:
home_data.head()
home_data.describe()
```

```python
home_data.columns
```

```python
# Idea: skip non-numeric data; use ~5 values (columns), skip data with missing values
# home_data = home_data.dropna(axis=0) # OK, this leaves us empty
# home_data.describe()

# try reducing the columns first: 
home_features = ['Id', 'MSSubClass', 'LotArea', 'OverallQual', 'OverallCond', 'YrSold']

X = home_data[home_features]
X.describe()
```

```python

# try again: 
X = X.dropna()
X.describe()
```

```python
X.shape
```

```python
# OK, same result. :-) 
# NB: Count is the same as in full set!
# Let's continue. 

# Define y: 
#y = home_data.SalePrice
y = home_data['SalePrice']
y.shape
```

```python
# Define model: 
from sklearn.ensemble import RandomForestRegressor
home_model = RandomForestRegressor(max_depth=5, random_state=0, n_estimators=100)

# Split into validation and training data
train_X, val_X, train_y, val_y = train_test_split(X, y, test_size=0.2, random_state=1)
 
# Fit model: 
home_model.fit(train_X, train_y) #X, y) 
```

```python
# First test: 
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")
print(home_model.predict(X.head()))
```

```python
# Make validation predictions: 
home_model_val_pred = home_model.predict(val_X)

# Calculate the mean absolute error of your Random Forest model on the validation data
home_val_mae = mean_absolute_error(home_model_val_pred, val_y)

print("Validation MAE for Random Forest Model: {}".format(home_val_mae))
```

```python
# So what's next? Compute predicted prices for ... which houses? Test data? 

#load data
test_file_path = "../data/Housing Prices Competition/test.csv"
test_data = pd.read_csv(test_file_path)

# make prediction? 
test_pred_prices = home_model.predict(test_data[home_features])

print(test_pred_prices)
# Tuples: Id, test_price
#for index in range(10):
#        print(test_data[index].ID,  test_data[index].SalePrice, test_pred_prices[index])
```
